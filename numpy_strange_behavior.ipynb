```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([-0.43716815, -0.43716815, 0.15707963])
alpha = np.array([0.40840704, 0.40840704, 1.02101761])

i = 0
fnc = lambda m: (x[i] - 2*np.pi*m) / alpha[i]
```

```python
x_axis = np.linspace(-0.2, 0.2, 100)
arr = x[i] - np.array([2*np.pi*mi for mi in x_axis]) / alpha[i]
arr2 = np.array([(x[i] - 2*np.pi*mi) / alpha[i] for mi in x_axis]) / alpha[i]
arr3 = np.array(list(map(fnc, x_axis)))
```

```python
sol1 = (x[i] - alpha[i])/(2*np.pi)
print(sol1, (x[i] - 2*np.pi*sol1)/alpha[i])

sol2 = (x[i] + alpha[i])/(2*np.pi)
print(sol2, (x[i] - 2*np.pi*sol2)/alpha[i])
```

```python
plt.figure(figsize=(20,8))

plt.plot(x_axis, arr, '-o')
plt.plot(x_axis, arr2, '-o')
plt.plot(x_axis, arr3, '-o')
#plt.plot(x_axis, [1 for _ in x_axis])
#plt.plot(x_axis, [-1 for _ in x_axis])

plt.vlines(sol1, ymin=np.min(arr2), ymax=np.max(arr2))
plt.vlines(sol2, ymin=np.min(arr2), ymax=np.max(arr2))
plt.hlines(xmin=np.min(x_axis), xmax=np.max(x_axis), y=(x[i] - 2*np.pi*sol1)/alpha[i])
plt.hlines(xmin=np.min(x_axis), xmax=np.max(x_axis), y=(x[i] - 2*np.pi*sol2)/alpha[i]);
```
